```python
dataset=pd.read_csv('train.csv')
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dataset,dataset['SalePrice'],test_size=0.1,random_state=0)
```

```python
categorical_feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=="O" ]

for feature in categorical_feature_nan:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(), 4)))
```

```python
categorical_feature_nan
```

```python
def replace_with_na(data,categorical_feature_nan):
    data=dataset.copy()
    data[categorical_feature_nan]=data[categorical_feature_nan].fillna("Missing")
    return data
dataset=replace_with_na(dataset,categorical_feature_nan)
dataset[categorical_feature_nan].isnull().sum()
```

```python
dataset
```

```python
numerical_values_with_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and feature not in categorical_feature_nan]

for feature in numerical_values_with_na:
    print("{}: {} % missing values".format(feature, np.round(dataset[feature].isnull().mean(),4)) )
```

```python
for feature in numerical_values_with_na:
    median_value=dataset[feature].median()
    dataset[feature+"nan"]=np.where(dataset[feature].isnull(),1,0)    
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numerical_values_with_na].isnull().sum()
dataset[numerical_values_with_na].isnull().sum()
```

```python
dataset
```

### Handling Temporal variable

```python
temporal_variable=["YearRemodAdd","YearBuilt","GarageYrBlt"]

for feature in temporal_variable:
    dataset[feature]=dataset["YrSold"]-dataset[feature]
```

```python
dataset[["GarageYrBlt","YearRemodAdd","YearBuilt"]]
```

```python
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes=="O" and feature not in ["Id"] ]
categorical_feature
```

```python
len(categorical_feature)
```

#### handling rare occurence in dataset

```python
temp=dataset.groupby("MSZoning")["SalePrice"].count()/len(dataset)
temp
```

```python
for feature in categorical_feature:
    temp=dataset.groupby(feature)["SalePrice"].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],"Rare_var")
    
```

```python
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
for feature in categorical_feature:
    labels_ordered=dataset.groupby(feature)["SalePrice"].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered)}
    dataset[feature]=dataset[feature].map(labels_ordered)
    


dataset.head(100)
```

```python
scaling_feature=[feature for feature in dataset.columns if feature not in ["Id","SalePrice"]]
len(scaling_feature)
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler.fit(dataset[scaling_feature])
```

```python
scaler.transform(dataset[scaling_feature])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[scaling_feature]), columns=scaling_feature)],
                    axis=1)

data
```
