#### Optimizers are a set of procedures defined in SciPy that either find the minimum value of a function, or the root of an equation.

```python
'CG''BFGS''Newton-CG''L-BFGS-B''TNC''COBYLA''SLSQP'

from scipy.optimize import minimize

def eqn(x):
  return x**2 + x + 2

mymin = minimize(eqn, 0, method='BFGS')
mymin2 = minimize(eqn, 0, method='CG')
mymin3 = minimize(eqn, 0, method='SLSQP')



print(mymin.x)
print(mymin2.x)
print(mymin3.x)
```

### Sparse Data: is a data set where most of the item values are zero.

####Dense Array: is the opposite of a sparse array: most of the values are not zero.

```python
import numpy as np
from scipy.sparse import csr_matrix

# (0,5)    1 means row 0 position 5 value is 1

arr = np.array([0, 0, 0, 0, 0, 1, 1, 0, 2])

print(csr_matrix(arr))
```

```python
import numpy as np
from scipy.sparse import csc_matrix

arr = np.array([[0, 0, 0], 
                [0, 0, 1],
                [1, 0, 2]])

print(csr_matrix(arr).data)
print(csc_matrix(arr))
print()
print(csr_matrix(arr))
```

#### Counting nonzeros with the count_nonzero() method:

```python
arr = np.array([[0, 0, 0], [0, 0, 1], [1, 0, 2]])

print(csr_matrix(arr).count_nonzero())
```

```python
arr = np.array([[0, 0, 0], [0, 0, 1], [1, 0, 2]])

mat = csr_matrix(arr)
mat.eliminate_zeros()

print(mat)
```


#### Eliminating duplicate entries with the sum_duplicates() method:


```python
arr = np.array([[0, 0, 0], [0, 0, 1], [1, 0, 2]])

mat = csr_matrix(arr)
mat.sum_duplicates()

print(mat)
```


#### Converting from csr to csc with the tocsc() method:

```python
arr = np.array([[0, 0, 0], [0, 0, 1], [1, 0, 2]])

newarr = csr_matrix(arr).tocsc()

print(newarr)
```

```python
from scipy.sparse.csgraph import connected_components
from scipy.sparse import csr_matrix

arr = np.array([
  [0, 1, 2],
  [1, 0, 0],
  [2, 0, 0]
])

newarr = csr_matrix(arr)

print(connected_components(newarr))
```

```python
import numpy as np
from scipy.sparse.csgraph import dijkstra
from scipy.sparse import csr_matrix

arr = np.array([
  [0, 1, 2],
  [1, 0, 0],
  [2, 0, 3]
])

newarr = csr_matrix(arr)

print(dijkstra(newarr, return_predecessors=False, indices=0))
print(dijkstra(newarr, return_predecessors=False, indices=2))
```
